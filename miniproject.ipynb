```python
import pandas as pd

df = pd.read_csv('kidney_disease_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

df['target'].value_counts()
```

```python
df.groupby('target')['age_of_the_patient'].mean()

df.groupby('target')['blood_pressure_(mm/hg)'].mean()

df.groupby('target')['body_mass_index_(bmi)'].mean()

df.groupby('target')['c-reactive_protein_(crp)_level'].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='target', y='age_of_the_patient', data=df)
plt.title('Age Distribution by Target')

sns.boxplot(x='target', y='blood_pressure_(mm/hg)', data=df)
plt.title('Blood Pressure by Target')

sns.violinplot(x='target', y='c-reactive_protein_(crp)_level', data=df)
plt.title('CRP Level by Target')
```

```python
df_binary = df[df['target'].isin(['Severe_Disease', 'No_Disease'])]

num_df = df.select_dtypes(include='number')
corr_matrix = num_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (All Numeric Features)')
```

```python
num_df = df.select_dtypes(include='number')
corr = num_df.corr().abs()
```

```python
import numpy as np
upper_tri = np.triu(np.ones(corr.shape), k=1).astype(bool)
top_corr = corr.where(upper_tri).unstack().dropna()
top_corr = top_corr.sort_values(ascending=False).head(10)

top_corr.plot(kind='barh', color='salmon')
plt.title('Top 10 Feature Correlations')
plt.xlabel('Correlation')
plt.grid()
plt.show()
```

```python
#In this analysis, I created an Inflammation Index by multiplying the values of CRP and IL-6.

df['inflammation_index'] = df['c-reactive_protein_(crp)_level'] * df['interleukin-6_(il-6)_level']
#I then categorized patients into two groups based on their Cystatin C levels,
#using a threshold of 1.0 mg/L — a commonly accepted clinical cutoff above which impaired kidney function is often suspected

#distribution of the inflammation index between the Normal and High Cystatin groups
plt.figure(figsize=(10, 5))
sns.boxplot(x='cystatin_risk_group', y='inflammation_index', data=df)
plt.title('Inflammation Index by Cystatin C Risk Group')
plt.show()
```
